### Task: Price Range Distribution

Let's visualize the distribution of price ranges and calculate the percentage of restaurants in each category.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram or bar chart to visualize the distribution of price ranges
if 'Price range' in df.columns:
    price_range_counts = df['Price range'].value_counts().sort_index()

    plt.figure(figsize=(8, 6))
    sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette='viridis')
    plt.title('Distribution of Price Ranges Among Restaurants')
    plt.xlabel('Price Range (1=cheap, 4=expensive)')
    plt.ylabel('Number of Restaurants')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    # Calculate the percentage of restaurants in each price range category
    total_restaurants = len(df)
    percentage_price_range = (price_range_counts / total_restaurants) * 100

    print("\nPercentage of Restaurants in Each Price Range Category:")
    display(percentage_price_range)
else:
    print("Column 'Price range' not found. Please check the dataset for the correct column name.")
```

### Task: Online Delivery

Now, let's analyze online delivery options and their impact on average ratings.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Dataset .csv')

# Display the first 5 rows and column information to understand the data structure
print(df.info())
display(df.head())
```

```python
# Determine the percentage of restaurants that offer online delivery
if 'Has Online delivery' in df.columns:
    online_delivery_counts = df['Has Online delivery'].value_counts()
    percentage_online_delivery = (online_delivery_counts / len(df)) * 100

    print("Percentage of Restaurants Offering Online Delivery:")
    display(percentage_online_delivery)

    # Compare the average ratings of restaurants with and without online delivery
    if 'Aggregate rating' in df.columns:
        avg_rating_by_delivery = df.groupby('Has Online delivery')['Aggregate rating'].mean()

        print("\nAverage Ratings of Restaurants With and Without Online Delivery:")
        display(avg_rating_by_delivery)

        plt.figure(figsize=(7, 5))
        sns.barplot(x=avg_rating_by_delivery.index, y=avg_rating_by_delivery.values, palette='coolwarm')
        plt.title('Average Rating by Online Delivery Availability')
        plt.xlabel('Has Online Delivery')
        plt.ylabel('Average Rating')
        plt.xticks([0, 1], ['No', 'Yes'])
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
    else:
        print("Column 'Aggregate rating' not found. Cannot compare average ratings.")
else:
    print("Column 'Has Online delivery' not found. Please check the dataset for the correct column name.")
```
